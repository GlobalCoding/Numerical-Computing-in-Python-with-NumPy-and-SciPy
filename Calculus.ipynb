## Differentiation

The ```scipy.misc.differentiate``` [function](https://docs.scipy.org/doc/scipy/reference/generated/scipy.misc.derivative.html) calcualtes the differential of a function at a given location by sampling values of the function around the location for which the derivative is requested.

<figcaption>Differentiation diagram</figcaption></center>

For instance, to estimate the first derivative, the calcualted result will be:

$ \frac{\textrm{d}y}{\textrm{d}x} =\frac{y(x+dx)-y(x-dx)}{2dx}$

When the ```scipy.misc.differentiate``` function is called, the first argument is the function that is to be differentiated (which must accept a single argument and return a scalar value). The second is the value of the independent variable which the derivative is to be calcualted at. It has a number of optional arguments which may be provided to describe how it should behave. ```order``` defines the number of points around the specified value of $x$ (this defaults to 3 if it is not specified and must be odd). ```dx``` specifies the spacing of these points (this defaults to 1). ```n``` specifies the number of times the function is to be differentiated before the differential is returned (this defaults to 1). The value of ```order``` must be greater than or equal to the value of ```n```.

Consider the function:

$f(x) = x^{3} + 3x^{2} - 2x +1$

We may calcualte the first, second and third derivatives:

$f'(x) = 3x^{2} + 6x -2\\
f''(x) = 6x + 6\\
f'''(x) = 6$

And we can evaluate these at specific values of x. For instance:

$f'(1) = 7\\
f''(2) = 18\\
f'''(0) = 6$

We can claulcate these values using SciPy.

````
from scipy.misc import derivative

def sample_function(x):
  return(x ** 3 + 3 * x ** 2 - 2 * x + 1)

# Note that dx defaults to 1, n defaults to 1 and order defaults to 3
print("First order: ", derivative(sample_function, 1, dx=1e-3))
print("Second order: ", derivative(sample_function, 2, n=2))
print("Third order: ", derivative(sample_function, 2, n=3, order = 5))
````

The results returned by this functions are senisitive to the value of ```dx```. Choosing too small a value leads to a round-off error which can invalidate the entire result. Choosing too large a value can lead to errors as sampling values far from the central value causes a resolution error. This means it may require some experiemntation to ensure the correct value is returned for the case in question. In general, higher derivatives require a larger value of ```dx```.

````
# Some examples when values of dx produce the wrong answer
print("First order: ", derivative(sample_function, 1))
print("Third order: ", derivative(sample_function, 2, n=3, order = 5, dx=1e-8))
````

In addition, the function may produce incorrect results or raise an exception at values where the function is discontinuous, undefined or infinite:

````
import math
from scipy.misc import derivative

# Differentiating at an inifinte value:
def one_over_x_squared(x):
  return(1 / x ** 2)

print(derivative(one_over_x_squared, 0))
````

````
from scipy.misc import derivative

# Differentiating a discountinuous function
def step(x):
  if x>0:
    return 1
  else:
    return 0

print("dx of 1: ", derivative(step, 0))
print("dx of 1e-8: ", derivative(step, 0, dx=1e-8))
````

````
import math
from scipy.misc import derivative

# Differentiating an undefined function:
print(derivative(math.log, -1))
````

All of these caveats means the function (like many in Numpy and Scipy) cannot be used indiscriminantly and without thought. So make sure you consider the case of cases the function will be used for in your applciation before applying it.

## Integration

Scipy contains a [suite to tools](https://docs.scipy.org/doc/scipy/reference/integrate.html) which can be used to integrate functions. Let's look at the ```scipy.integrate.quad``` [function](https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.quad.html#scipy.integrate.quad). Again, we'll use the function:

$f(x) = x^{3} + 3x^{2} - 2x +1$

When we integrate this, we obtain:

$\int f(x)\textrm{d}x = \frac{x^{4}}{4} + x^{3} - x^{2} + x$

We can evaluate this integral over a fixed range and find:

$\int\limits_{0}^{1} f(x)\textrm{d}x = \frac{5}{4}$

We can call ```scipy.integrate.quad``` to evaluate this value. The first argument is the function (which must accept a single argument and return a scalar value). The second and third values are the upper and lower bounds of the definite integral.

````
from scipy.integrate import quad

def sample_function(x):
  return(x ** 3 + 3 * x ** 2 - 2 * x + 1)

print(quad(sample_function, 0, 1))
````

Here, ```quad``` has returned a tuple containing two values. The first of these is the value of integral as evalauted by quad, the second is the estimate of the maximum error of the function.
